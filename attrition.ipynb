## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras import layers
from tensorflow.keras.layers import Dense, Input

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns

y_df = attrition_df[['Attrition' , 'Department']]
y_df.head()
```

```python
# looking at the total number of response options for department
dept_counts = y_df['Department'].value_counts()
dept_counts
```

```python
# Create a list of at least 10 column names to use as X data
x_columns = ["JobLevel",
    "DistanceFromHome",
    "JobInvolvement",
    "JobSatisfaction",
    "PerformanceRating",
    "OverTime",
    "StockOptionLevel",
    "HourlyRate",
    "TrainingTimesLastYear",
    "WorkLifeBalance",
    "YearsWithCurrManager",
    "YearsSinceLastPromotion"]

# Create X_df using your selected columns
X_df = attrition_df[x_columns]

# Show the data types for X_df
X_df.info()
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, random_state=42)
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
X_train['OverTime'] = X_train['OverTime'].apply(lambda x: 1 if x == 'Yes' else 0).astype('int64')
X_test['OverTime'] = X_test['OverTime'].apply(lambda x: 1 if x == 'Yes' else 0).astype('int64')

# Print the resulting Series
overtime_counts = X_test['OverTime'].value_counts()
print(overtime_counts)
```

```python
# Mine: Just looking at the full x_train to make certain I didn't miss anything
print(X_train.info())
print(X_test.info())
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
# Create a StandardScaler
stand_scaler = StandardScaler()

# Fit the StandardScaler to the training data
stand_scaler.fit(X_train)

# Scale the training and testing data
X_train_stand_scaled = stand_scaler.transform(X_train)
X_test_stand_scaled = stand_scaler.transform(X_test)
```

```python
# Create a OneHotEncoder for the Department column
enc = OneHotEncoder(sparse_output=False, dtype='int')

# Fit the encoder to the training data
enc.fit(y_train[['Department']])

# Create two new variables by applying the encoder
# to the training and testing data
y_enc_dept_train = enc.transform(y_train[['Department']])
y_enc_dept_test = enc.transform(y_test[['Department']])

print(y_enc_dept_test)
print(y_enc_dept_train.shape)
print(y_enc_dept_test.shape)
```

```python
# Create a OneHotEncoder for the Attrition column
# Fit the encoder to the training data
y_attrition = enc.fit(y_train[['Attrition']])

# Create two new variables by applying the encoder
# to the training and testing data
y_enc_att_train = y_attrition.transform(y_train[['Attrition']])
y_enc_att_test = y_attrition.transform(y_test[['Attrition']])

y_enc_att_test[:5]
```

# Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
number_columns = X_train_stand_scaled.shape[1]

# Create the input layer
input_layer = Input(shape=(number_columns,), name = 'input')

# Create at least two shared layers
shared_layer1 = Dense(64, activation='relu')(input_layer)
shared_layer2 = Dense(128, activation='relu')(shared_layer1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer
# Create the hidden layer
dept_hidden_layer = layers.Dense(32, activation='relu')(shared_layer2)

# Create the output layer
department_output = layers.Dense(3, activation='softmax', name='Department')(dept_hidden_layer)
```

```python
# Create a branch for Attrition
# Create the hidden layer
attrition_hidden_layer = layers.Dense(32, activation='relu')(shared_layer2)

# Create the output layer
attrition_output = layers.Dense(2, activation='sigmoid', name='Attrition')(attrition_hidden_layer)
```

```python
# Create the model
nn_model = Model(inputs=input_layer, outputs=[department_output, attrition_output])

# Compile the model
nn_model.compile(optimizer="adam",
    loss={'Department' :'categorical_crossentropy',
          'Attrition' : 'binary_crossentropy'} ,
    metrics=['accuracy'])

# Summarize the model
nn_model.summary()
```

```python
# Train the model
trained_model = nn_model.fit(x= X_train_stand_scaled,
             y={'Department': y_enc_dept_train, 'Attrition': y_enc_att_train},
             epochs=100,
             validation_data=(X_test_stand_scaled, {'Department': y_enc_dept_test, 'Attrition': y_enc_att_test}))
```

```python
# Evaluate the model with the testing data
results = nn_model.evaluate(X_test_stand_scaled,
    {'Department': y_enc_dept_test,
     'Attrition': y_enc_att_test})
```

```python
# Print the accuracy for both department and attrition
print(f"Department predictions accuracy: {results[3]}")
print(f"Attrition predictions accuracy: {results[4]}")
```
